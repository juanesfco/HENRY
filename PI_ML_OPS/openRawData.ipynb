## User Reviews file

### Reading data

First, decompress the file. Note that the raw data uses single quotes (') instead of double quotes (") for the string, this is not recognized as json. So an initial transformation had to be made to open it.

```python
!gzip -dk RawData/user_reviews.json.gz
```

```python
import json
data = []
with open('RawData/user_reviews.json','r') as f:
    for l in f:
        l = l.replace("\"", "\'")
        l = l.replace("\'user_id\': \'", "\"user_id\": \"")
        l = l.replace("\', \'user_url\': \'", "\", \"user_url\": \"")
        l = l.replace("\', \'reviews\': [", "\", \"reviews\": \"[")
        l = l.replace("]}", "]\"}")
        l = l.replace("\\", "")
        
        data.append(json.loads(l))
data[0:2]
```

### Convert to Dataframe and Save as .csv

```python
import pandas as pd
df_reviews = pd.DataFrame(data)
df_reviews
```

```python
df_reviews.to_csv('Data/df_reviews.csv',index=False)
```

## User Items file

### Reading data

First, decompress the file. Note that the raw data uses single quotes (') instead of double quotes (") for the string, this is not recognized as json. So an initial transformation had to be made to open it.

```python
!gzip -dk RawData/users_items.json.gz
```

```python
import json
data = []
with open('RawData/users_items.json','r') as f:
    for l in f:
        l = l.replace("\"", "\'")
        l = l.replace("\'user_id\': \'", "\"user_id\": \"")
        l = l.replace("\', \'items_count\': ", "\", \"items_count\": \"")
        l = l.replace(", \'steam_id\': \'", "\", \"steam_id\": \"")
        l = l.replace("\', \'user_url\': \'", "\", \"user_url\": \"")
        l = l.replace("\', \'items\': [", "\", \"items\": \"[")
        l = l.replace("]}", "]\"}")
        l = l.replace("\\", "")
        
        data.append(json.loads(l))
data[0]
```

### Convert to Dataframe and Save as .csv

```python
import pandas as pd
df_user_items = pd.DataFrame(data)
df_user_items
```

```python
df_user_items.to_csv('Data/df_user_items.csv',index=False)
```
